```python
import requests
from sklearn.preprocessing import StandardScaler

from bs4 import BeautifulSoup
import csv
import re

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import (accuracy_score, recall_score, 
                           confusion_matrix, classification_report, 
                           roc_auc_score)

import shap

import sweetviz as sv

import optuna
from catboost import CatBoostClassifier
from concurrent.futures import ThreadPoolExecutor
```

```python


headers = {'User-Agent': 'Mozilla/5.0'}
main_url = "https://ratcatcher.ru/media/summer_prac/parcing/7/index.html"

def fetch_html(url):
    try: return requests.get(url, headers=headers, timeout=10).text
    except: return None

def parse_links():
    html = fetch_html(main_url)
    soup = BeautifulSoup(html, 'html.parser')
    return [{'name': a.text.strip(), 'url': a['href']} for a in soup.select('ul li a')]

def parse_page(link):
    html = fetch_html(link['url'])
    soup = BeautifulSoup(html, 'html.parser')
    table = soup.find('table', id='rice')
    if not table: return []
    headers = [th.text.strip() for th in table.select('thead th')]
    return [dict(zip(headers, [td.text.strip() for td in tr.select('td')])) for tr in table.select('tbody tr')]

links = parse_links()
with ThreadPoolExecutor(max_workers=8) as executor:
    data = [item for sublist in executor.map(parse_page, links) for item in sublist]

df = pd.DataFrame(data)  # Готовая таблица в переменной "df"
```

```python

def add_nonlinear_features(df):

    df = df.copy()
    df['Area_squared'] = df['Area'] ** 2
    df['Perimeter_squared'] = df['Perimeter'] ** 2
    df['MajorMinor_product'] = df['MajorAxisLength'] * df['MinorAxisLength']
    df['Area_roundness'] = df['Area'] * df['Roundness']

    df['Eccentricity_Aspect'] = df['Eccentricity'] * df['AspectRation']

    df['Log_Major'] = np.log(df['MajorAxisLength'] + 1)
    df['Inv_Minor'] = 1. / (df['MinorAxisLength'] + 1e-6)
    df['Log_Area'] = np.log(df['Area'] + 1)
    df['Sqrt_ConvexArea'] = np.sqrt(df['ConvexArea'])

    df['Exp_Extent'] = np.exp(df['Extent'])
    df['Exp_Roundness'] = np.exp(df['Roundness'])

    df['Sqrt_AreaPerimeter_ratio'] = np.sqrt(df['Area'] / (df['Perimeter'] + 1e-6))

    df['Sin_Aspect'] = np.sin(df['AspectRation'])
    df['Cos_Aspect'] = np.cos(df['AspectRation'])

    df['Cube_MajorPlusMinor'] = (df['MajorAxisLength'] + df['MinorAxisLength']) ** 3
    df['Log_Area_times_Aspect'] = np.log(df['Area'] * df['AspectRation'] + 1)

    return df
```

```python
def select_correlated_features(df, target_col='Class', threshold=0.4):
    '''Отбор признаков с высокой корреляцией с целевой переменной'''
    numeric_cols = df.select_dtypes(include=[np.number]).columns.drop([target_col])
    correlations = df[numeric_cols].corrwith(df[target_col]).abs()
    selected_features = correlations[correlations >= threshold.index.tolist()]
    return selected_features, correlations.sort_values(ascending=False)
```

```python
def select_high_correlation_features(df, target_col='Class', threshold=0.4):

    corrs = df.corr(numeric_only=True)[target_col].drop(target_col).abs()
    selected = corrs[corrs >= threshold].index.tolist()
    selected.append(target_col)
    return df[selected]
```

```python
df_scaled_clean=select_high_correlation_features(df_scaled)
```

```python
from catboost import CatBoostClassifier
from sklearn.model_selection import train_test_split, GridSearchCV

X = df_scaled_clean.drop(columns=['Class'])
y = df_scaled_clean['Class']
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

param_grid = {
    'depth': [4, 6, 8],
    'iterations': [300, 500, 1000],
    'learning_rate': [0.01, 0.05, 0.1],
    'l2_leaf_reg': [1, 3, 5]
}

cbc = CatBoostClassifier(verbose=0, random_seed=42)

grid_search = GridSearchCV(
    estimator=cbc,
    param_grid=param_grid,
    scoring='accuracy',
    cv=3,
    n_jobs=-1
)

grid_search.fit(X_train, y_train)

print("Лучшие параметры:", grid_search.best_params_)
print("Лучшая точность (кросс-валидация): {:.4f}".format(grid_search.best_score_))

best_model = grid_search.best_estimator_
```

```python
from sklearn.metrics import accuracy_score, recall_score, confusion_matrix

def print_model_metrics(model, X_test, y_test):
    """
    Выводит accuracy, recall и confusion matrix (матрицу ошибок) для классификатора.

    Аргументы:
        model: обученная модель CatBoostClassifier (или аналогичная).
        X_test: тестовые признаки.
        y_test: истинные метки классов.
    """
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    rec = recall_score(y_test, y_pred, average="macro")

    print(f'Accuracy: {acc:.4f}')
    print(f'Recall: {rec:.4f}')

```

```python
print_model_metrics(best_model, X_test, y_test)
```
